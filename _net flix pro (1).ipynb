```python
# Data Preprocessing:
# Handle Missing Values
# Fill missing values with 'Unknown'
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['date_added'].fillna(method='ffill', inplace=True)  # optional

# Confirm no missing values remain
df.isnull().sum()
```

```python
# Convert Categorical Data into Numeric Format
# Extract numeric duration and its unit (e.g., minutes, seasons)
df['duration'] = df['duration'].fillna('0')
df['duration_int'] = df['duration'].str.extract('(\d+)').astype(int)
df['duration_type'] = df['duration'].str.extract('([a-zA-Z ]+)').fillna('Unknown')

# One-hot encode 'type', 'rating', and 'duration_type'
encoded_df = pd.get_dummies(df[['type', 'rating', 'duration_type']], drop_first=True)
```

```python
#  Standardize Numerical Features
from datetime import datetime
from sklearn.preprocessing import StandardScaler

# Create content age feature
df['content_age'] = datetime.now().year - df['release_year']

# Select numeric columns
numeric_cols = df[['duration_int', 'release_year', 'content_age']]

# Standardize
scaler = StandardScaler()
numeric_scaled = pd.DataFrame(scaler.fit_transform(numeric_cols), columns=numeric_cols.columns)
```

```python
# . TF-IDF Vectorization of listed_in (Genres)
from sklearn.feature_extraction.text import TfidfVectorizer

# Fill NaN and apply TF-IDF
tfidf = TfidfVectorizer(stop_words='english', max_features=20)
tfidf_matrix = tfidf.fit_transform(df['listed_in'].fillna(''))

# Convert to DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

```python
# Feature Engineering:
# Feature: Content Age
from datetime import datetime

# Current year
current_year = datetime.now().year

# New feature: content age
df['content_age'] = current_year - df['release_year']
```

```python
# Feature: Genre Count
# New feature: genre count
df['genre_count'] = df['listed_in'].apply(lambda x: len(str(x).split(',')))
```

```python
#  Transform Categorical Variables
# One-hot encode type, rating, and duration_type
encoded_df = pd.get_dummies(df[['type', 'rating', 'duration_type']], drop_first=True)
```

```python
# Assemble Final Engineered Dataset
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer

# Standardize numeric features
num_features = df[['duration_int', 'content_age', 'genre_count']]
scaler = StandardScaler()
num_scaled = pd.DataFrame(scaler.fit_transform(num_features), columns=num_features.columns)

# TF-IDF on 'listed_in' genres
tfidf = TfidfVectorizer(stop_words='english', max_features=20)
tfidf_matrix = tfidf.fit_transform(df['listed_in'].fillna(''))
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())

# Final dataset for clustering
final_df = pd.concat([num_scaled, encoded_df, tfidf_df], axis=1)
final_df.head()
```

```python
# Clustering Model Selection:
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
inertia = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(final_df)
    inertia.append(kmeans.inertia_)

# Plot the Elbow
plt.figure(figsize=(8,5))
plt.plot(k_range, inertia, marker='o')
plt.title("Elbow Method - Optimal K")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
#K-Means Clustering + Silhouette Score
# Let’s assume k=4 (you can adjust based on elbow curve)
kmeans = KMeans(n_clusters=4, random_state=42)
clusters_kmeans = kmeans.fit_predict(final_df)

# Add to original DataFrame
df['cluster_kmeans'] = clusters_kmeans

# Silhouette Score
score = silhouette_score(final_df, clusters_kmeans)
print("Silhouette Score for K-Means:", round(score, 3))
```

```python
# Hierarchical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Linkage matrix using Ward's method
linkage_matrix = linkage(final_df.sample(300), method='ward')  # subset if slow

# Plot dendrogram
plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix, truncate_mode='lastp', p=20)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Cluster Index")
plt.ylabel("Distance")
plt.show()
```

```python
# DBSCAN (Density-Based Clustering)
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1.2, min_samples=5)
clusters_dbscan = dbscan.fit_predict(final_df)

# Add to DataFrame
df['cluster_dbscan'] = clusters_dbscan

# Check unique labels
print("DBSCAN Clusters:", set(clusters_dbscan))
```

```python
# K-Means Clustering for Multiple k-values
# Loop Through Multiple k Values
# Define range of k values to test
k_values = range(2, 11)

# Store scores
inertias = []
silhouette_scores = []

# Run KMeans for each k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    preds = kmeans.fit_predict(final_df)
    
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(final_df, preds))

# Show results
for k, sil, inert in zip(k_values, silhouette_scores, inertias):
    print(f"K = {k} → Silhouette Score: {round(sil, 3)}, Inertia: {round(inert, 2)}")
```

```python
# Plot Elbow Curve & Silhouette Score
# Elbow Curve (Inertia)
plt.figure(figsize=(10,5))
plt.plot(k_values, inertias, marker='o')
plt.title("Elbow Method - Inertia vs K")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
#  Silhouette Score vs k
plt.figure(figsize=(10,5))
plt.plot(k_values, silhouette_scores, marker='o', color='green')
plt.title("Silhouette Score vs Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

```python
# Step 4A: Final K-Means Model Training (Assume best k = 4)
from sklearn.cluster import KMeans

# Final model with chosen k
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster_kmeans'] = kmeans.fit_predict(final_df)
```

```python
#  Step 4B: Hierarchical Clustering (With Linkage Comparison)
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Compute linkage matrix (method = 'ward', 'complete', 'average', 'single')
Z = linkage(final_df.sample(300), method='ward')  # sample if data is large

# Dendrogram
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=20)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Cluster")
plt.ylabel("Distance")
plt.show()

# Assign clusters (e.g., cut into 4)
clusters_hier = fcluster(Z, 4, criterion='maxclust')
```

```python
# Step 4C: DBSCAN Training
from sklearn.cluster import DBSCAN

# Train DBSCAN
dbscan = DBSCAN(eps=1.2, min_samples=5)
df['cluster_dbscan'] = dbscan.fit_predict(final_df)

# Unique clusters (note: -1 means noise)
print("DBSCAN clusters found:", set(df['cluster_dbscan']))
```

```python
# Step 4D: Dimensionality Reduction for Visualization (PCA & t-SNE)
# PCA (Principal Component Analysis)
from sklearn.decomposition import PCA
import seaborn as sns

# Reduce to 2 dimensions
pca = PCA(n_components=2)
reduced_pca = pca.fit_transform(final_df)

# Add PCA columns to DataFrame
df['pca1'] = reduced_pca[:, 0]
df['pca2'] = reduced_pca[:, 1]

# Plot PCA clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='pca1', y='pca2', hue='cluster_kmeans', data=df, palette='tab10')
plt.title("K-Means Clusters (PCA Reduced)")
plt.show()
```

```python
#  t-SNE (Optional — great for nonlinear separation)
from sklearn.manifold import TSNE

# Reduce to 2D using t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
reduced_tsne = tsne.fit_transform(final_df)

# Add t-SNE coordinates
df['tsne1'] = reduced_tsne[:, 0]
df['tsne2'] = reduced_tsne[:, 1]

# Plot t-SNE clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tsne1', y='tsne2', hue='cluster_kmeans', data=df, palette='tab10')
plt.title("K-Means Clusters (t-SNE Reduced)")
plt.show()
```

```python
# Step 5: Visualization & Interpretation
#  Cluster Visualization Using PCA
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(x='pca1', y='pca2', hue='cluster_kmeans', data=df, palette='tab10')
plt.title("K-Means Clusters (PCA 2D Projection)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()
```

```python
# Create a cluster-aggregated version of selected features
cluster_profile = df.groupby('cluster_kmeans')[['duration_int', 'content_age', 'genre_count']].mean()

# Heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(cluster_profile, annot=True, cmap='coolwarm')
plt.title("Mean Feature Values per Cluster (K-Means)")
plt.show()
```

```python
# Most Common Genres per Cluster
# Add all genres to a new column list
df['genres_list'] = df['listed_in'].apply(lambda x: [genre.strip() for genre in str(x).split(',')])

# Explode genres for analysis
df_exploded = df.explode('genres_list')

# Top genres by cluster
top_genres = df_exploded.groupby('cluster_kmeans')['genres_list'].value_counts().groupby(level=0).nlargest(3)
print("Top 3 Genres per Cluster:\n")
print(top_genres)
```

```python
# Rating Distribution by Cluster
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='rating', hue='cluster_kmeans', order=df['rating'].value_counts().index[:10], palette='tab10')
plt.title("Content Ratings Distribution by Cluster")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Cluster")
plt.show()
```

```python
# Evaluation & Refinement
#  Evaluate Your Final K-Means Model
from sklearn.metrics import silhouette_score, davies_bouldin_score

# KMeans with k=4 (or your best value)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(final_df)

# Evaluation metrics
sil_score = silhouette_score(final_df, kmeans_labels)
db_score = davies_bouldin_score(final_df, kmeans_labels)
inertia = kmeans.inertia_

print("✅ Silhouette Score:", round(sil_score, 3))
print("✅ Davies-Bouldin Index:", round(db_score, 3))
print("✅ Inertia:", round(inertia, 2))
```
